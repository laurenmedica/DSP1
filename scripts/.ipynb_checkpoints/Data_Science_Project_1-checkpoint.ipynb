Combine the `fake_df` and `true_df` DataFrames, add a 'label' column (0 for fake, 1 for true), and shuffle the combined dataset to ensure randomness. Then, handle any missing values in the text and title columns.

**Reasoning**:
Following the instructions, I will now add a 'label' column to the `true_df` DataFrame, setting its value to 1 to identify true news.



```python
true_df['label'] = 1
print("Added 'label' column to real_df with value 1.")
```

**Reasoning**:
Now that both `fake_df` and `true_df` have their respective 'label' columns, I will concatenate them vertically into a new DataFrame called `combined_df`.



```python
combined_df = pd.concat([fake_df, true_df], ignore_index=True)
print("DataFrames combined into 'combined_df'.")
```

**Reasoning**:
As per the instructions, I will now randomly shuffle the rows of the `combined_df` DataFrame to ensure randomness and then reset its index.



```python
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)
print("Combined DataFrame shuffled and index reset.")
```

**Reasoning**:

**Reasoning**:
Next, I will check for and handle any missing values in the 'text' column of the `combined_df` by filling them with an empty string, as specified in the instructions.

## Preprocess Text

### Subtask:
Clean and preprocess the 'title' and 'text' columns by converting text to lowercase, removing punctuation, stop words, and applying stemming or lemmatization to prepare them for vectorization.


**Reasoning**:
Before implementing the preprocessing function, I need to ensure that the necessary NLTK components, specifically the 'stopwords' corpus, are downloaded. This is a one-time setup step for NLTK.



```python
import nltk
nltk.download('stopwords')
print("NLTK 'stopwords' corpus downloaded.")
```

**Reasoning**:
Now that the NLTK 'stopwords' corpus is downloaded, I will define a preprocessing function that converts text to lowercase, removes punctuation, tokenizes, removes stopwords, and applies stemming, and then apply it to the 'title' and 'text' columns of `combined_df` to create new processed columns as per the instructions.



```python
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk

# Ensure punkt tokenizer and punkt_tab are downloaded
try:
    word_tokenize('test')
except LookupError:
    nltk.download('punkt')
    nltk.download('punkt_tab') # Added explicit download for 'punkt_tab'
    print("NLTK 'punkt' and 'punkt_tab' resources downloaded.")

# Initialize stemmer and stopwords
stemmer = PorterStemmer() # Corrected typo: PorterStemter to PorterStemmer
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = str(text).lower()  # Convert to string and lowercase
    text = ''.join([char for char in text if char not in string.punctuation]) # Remove punctuation
    tokens = word_tokenize(text) # Tokenize text
    tokens = [word for word in tokens if word not in stop_words] # Remove stopwords
    tokens = [stemmer.stem(word) for word in tokens] # Apply stemming
    return ' '.join(tokens)

# Apply preprocessing to 'title' and 'text' columns
combined_df['processed_title'] = combined_df['title'].apply(preprocess_text)
combined_df['processed_text'] = combined_df['text'].apply(preprocess_text)

print("Title and text columns preprocessed and new 'processed_title' and 'processed_text' columns created.")
```

**Reasoning**:
To verify that the preprocessing steps were applied correctly and to inspect the newly created `processed_title` and `processed_text` columns, I will display the first few rows of the `combined_df` DataFrame.



```python
display(combined_df.head())
```

## Vectorize Text Features

### Subtask:
Convert the preprocessed 'title' and 'text' columns into numerical features using TF-IDF Vectorization. This will create two sets of features: one for titles (`X_title`) and one for full text (`X_text`). Split both feature sets, along with the 'label' column, into training and testing sets.


**Reasoning**:
To prepare the text data for machine learning, I will first import the necessary modules for TF-IDF vectorization and data splitting. Then, I will initialize and apply TF-IDF vectorization separately to the 'processed_title' and 'processed_text' columns, and finally split these vectorized features and the 'label' column into training and testing sets, ensuring consistent splits.



```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Initialize TfidfVectorizer for 'processed_title'
tfidf_vectorizer_title = TfidfVectorizer(max_features=5000) # Limiting features to manage dimensionality
X_title = tfidf_vectorizer_title.fit_transform(combined_df['processed_title'])
print("TF-IDF vectorization for 'processed_title' completed. Shape of X_title:", X_title.shape)

# Initialize TfidfVectorizer for 'processed_text'
tfidf_vectorizer_text = TfidfVectorizer(max_features=10000) # Limiting features to manage dimensionality
X_text = tfidf_vectorizer_text.fit_transform(combined_df['processed_text'])
print("TF-IDF vectorization for 'processed_text' completed. Shape of X_text:", X_text.shape)

# Define the target variable
y = combined_df['label']

# Split X_title and y into training and testing sets
X_train_title, X_test_title, y_train, y_test = train_test_split(X_title, y, test_size=0.2, random_state=42)
print("X_title and y split into training and testing sets.")
print(f"X_train_title shape: {X_train_title.shape}, X_test_title shape: {X_test_title.shape}")
print(f"y_train shape: {y_train.shape}, y_test shape: {y_test.shape}")

# Split X_text into training and testing sets (using the same y_train and y_test)
X_train_text, X_test_text, _, _ = train_test_split(X_text, y, test_size=0.2, random_state=42)
print("X_text split into training and testing sets (using the same y splits).")
print(f"X_train_text shape: {X_train_text.shape}, X_test_text shape: {X_test_text.shape}")
```

#### After we processed all the data the we will put the the copy of the final data into the Data folder

```python
combined_df.to_csv("../data/final_data.csv", index=False)
```

## Train and Evaluate Random Forest Model on Titles

### Subtask:
Train a Random Forest Classifier using the TF-IDF vectorized 'title' features (X_train_title) and the corresponding labels (y_train). Evaluate the model's performance on the test set (X_test_title, y_test) by calculating and displaying the accuracy, classification report, and confusion matrix.


**Reasoning**:
To begin training and evaluation, I will first import the necessary modules: `RandomForestClassifier` for the model, and `accuracy_score`, `classification_report`, and `confusion_matrix` for evaluating its performance.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Necessary modules imported for Random Forest classification and evaluation.")
```

**Reasoning**:
Now that the necessary modules are imported, I will initialize a RandomForestClassifier model, train it with the training data (X_train_title, y_train), make predictions on the test data (X_test_title), and then calculate and display the accuracy score, classification report, and confusion matrix to evaluate its performance.



```python
model_title_rf = RandomForestClassifier(random_state=42)
model_title_rf.fit(X_train_title, y_train)

y_pred_title_rf = model_title_rf.predict(X_test_title)

accuracy_title_rf = accuracy_score(y_test, y_pred_title_rf)
print(f"Accuracy for Random Forest on Titles: {accuracy_title_rf:.4f}\n")

print("Classification Report for Random Forest on Titles:")
classification_report_titles = classification_report(y_test, y_pred_title_rf, digits=4)
print(classification_report_titles)

print("Confusion Matrix for Random Forest on Titles:")
confusion_matrix_titles = confusion_matrix(y_test, y_pred_title_rf)
print(confusion_matrix_titles)
```

```python
# Writing results to the output file
with open("../output/random_forest_title_results.txt", "w") as f:
    f.write(classification_report_titles)
    f.write("\n\nConfusion Matrix:\n")
    f.write(str(confusion_matrix_titles))
```

## Train and Evaluate Random Forest Model on Full Text

### Subtask:
Train a separate Random Forest Classifier using the TF-IDF vectorized 'full text' features (X_train_text) and the corresponding labels (y_train). Evaluate this model's performance on its respective test set (X_test_text, y_test) by calculating and displaying the accuracy, classification report, and confusion matrix.


**Reasoning**:
Now that the necessary modules are imported, I will initialize a RandomForestClassifier model, train it with the training data (X_train_text, y_train), make predictions on the test data (X_test_text), and then calculate and display the accuracy score, classification report, and confusion matrix to evaluate its performance on the full text features.



```python
model_text_rf = RandomForestClassifier(random_state=42)
model_text_rf.fit(X_train_text, y_train)

y_pred_text_rf = model_text_rf.predict(X_test_text)

accuracy_text_rf = accuracy_score(y_test, y_pred_text_rf)
print(f"Accuracy for Random Forest on Text: {accuracy_text_rf:.4f}\n")

print("Classification Report for Random Forest on Text:")
classification_report_text = classification_report(y_test, y_pred_text_rf, digits=4)
print(classification_report_text)

print("Confusion Matrix for Random Forest on Text:")
confusion_matrix_text = confusion_matrix(y_test, y_pred_text_rf)
print(confusion_matrix_text)
```

```python
# Writing results to the output file
with open("../output/random_forest_text_results.txt", "w") as f:
    f.write(classification_report_text)
    f.write("\n\nConfusion Matrix:\n")
    f.write(str(confusion_matrix_text))
```

## Compare Model Performance

### Subtask:
Compare the performance metrics (especially accuracy) of the two Random Forest models (title-based vs. full text-based) and create a visualization (e.g., a bar chart) to clearly illustrate their differences. Discuss the findings in relation to the research question.


**Reasoning**:
To numerically compare the performance of the two Random Forest models, I will first print their respective accuracy scores, which were previously calculated as `accuracy_title_rf` and `accuracy_text_rf`.



```python
print(f"Accuracy of Random Forest on Titles: {accuracy_title_rf:.4f}")
print(f"Accuracy of Random Forest on Text: {accuracy_text_rf:.4f}")
```

**Reasoning**:
To visualize the comparison of the two models' accuracies, I will import the `matplotlib.pyplot` module, which is essential for creating plots in Python.



```python
import matplotlib.pyplot as plt
print("matplotlib.pyplot imported as plt.")
```

**Reasoning**:
Now that `matplotlib.pyplot` is imported, I will create a bar chart to visually compare the accuracy of the Random Forest model trained on titles and the one trained on full text, as specified in the instructions. This involves defining the data, creating the plot, adding labels and a title, and then displaying it.



```python
fig, ax = plt.subplots(figsize=(8, 6))

models = ['Title Model', 'Text Model']
accuracies = [accuracy_title_rf, accuracy_text_rf]

ax.bar(models, accuracies, color=['skyblue', 'lightcoral'])
ax.set_ylim(0.8, 1.0)
ax.set_ylabel('Accuracy')
ax.set_title('Comparison of Random Forest Model Accuracies')
ax.grid(axis='y', linestyle='--', alpha=0.7)

for i, acc in enumerate(accuracies):
    ax.text(i, acc - 0.01, f'{acc:.4f}',
            ha='center', va='top')

# ✅ Save using the figure object
fig.savefig("../output/Comparison_of_Random_Forest_Model_Accuracies.pdf",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

### Discussion of Model Performance and Research Question

The Random Forest model trained on the full text data (`accuracy_text_rf = 0.9959`) significantly outperformed the model trained solely on the titles (`accuracy_title_rf = 0.9400`). This indicates that the complete content of the news articles provides much richer and more discriminative features for classifying fake news compared to just the titles.

Regarding the research question about the amount of text needed for accurate prediction, it is clear that utilizing the full text is crucial for achieving higher accuracy in this specific fake news detection task. While titles alone can provide a decent level of accuracy (94%), the additional context, details, and linguistic patterns found in the full article allow the model to make more informed and precise classifications, leading to nearly perfect accuracy.

Both models successfully met and far exceeded the 90% accuracy goal. The title-based model achieved 94% accuracy, and the full text-based model achieved an outstanding 99.59% accuracy.
